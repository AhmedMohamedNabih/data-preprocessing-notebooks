```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.impute import KNNImputer, SimpleImputer
```

```python
knn=KNNImputer(n_neighbors=5)
df['BuildingArea']=knn.fit_transform(df[['BuildingArea']])
df['BuildingArea']
```

```python
# لو عايز احط قيمه معينه استخدم fill_value
sim_impu=SimpleImputer(strategy='median')
df[['Car','YearBuilt']]=sim_impu.fit_transform(df[['Car','YearBuilt']])
df[['Car','YearBuilt']]
```

```python
df.isna().sum()
```

```python
df_cat=df.select_dtypes(include='O').drop(['Suburb','Address','SellerG','Date','CouncilArea'],axis=1)
df_cat
```

```python
df.Type.value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder =OrdinalEncoder(categories=[['u','t','h']])
df['Type']=encoder.fit_transform(df_cat[['Type']])
```

```python
pd.get_dummies(df,columns=['Method'],drop_first=True)
```

```python
# لازم لما اعمل one hotencoder-> عشان اضيف الصفوف دي علي الداتا بتاعتي المفروض احولها الي داتا frame عشان اضفها 
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(drop='first',sparse_output=False)
ohe_arr=ohe.fit_transform(df[['Method']])
```

```python
# columns=ohe.get_feature_names_out() -> عشان اسماء الكولوم تطلع
df_ohe=pd.DataFrame(ohe_arr,columns=ohe.get_feature_names_out())
```

```python
df_cat=pd.concat([df_cat,df_ohe],axis=1).drop(['Method'],axis=1)
```
