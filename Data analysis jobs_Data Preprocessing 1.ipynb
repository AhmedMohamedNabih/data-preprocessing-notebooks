## feature Enginering 

```python
df.head()
```

```python
df.Size.unique()
```

```python
def size_new(x):
    if x in['1 to 50 employees','51 to 200 employees']:
        return 'startup company'
    elif x in ['201 to 500 employees','501 to 1000 employees',]:
        return 'median company'
    elif x in ['1001 to 5000 employees','5001 to 10000 employees','10000+ employees']:
        return 'large company'
    else:
        return x
df['size_categories']=df['Size'].apply(size_new)
df['size_categories']
```

```python
df['Revenue'].unique()
```

```python
def rev_categories(x):
    if type(x) == float:
        return x
    if x in 'Less than $1 million (USD)':
        return 'Low revenu company'
    elif x in ['$1 to $5 million (USD)',
    '$5 to $10 million (USD)',
    '$10 to $25 million (USD)',
    '$25 to $50 million (USD)',
    '$50 to $100 million (USD)']:
        return 'Menidan revenue company'
    elif x in ['$100 to $500 million (USD)',
    '$500 million to $1 billion (USD)']:
        return 'high revenue company'
    elif x in ['$1 to $2 billion (USD)',
    '$2 to $5 billion (USD)',
    '$5 to $10 billion (USD)',
    '$10+ billion (USD)']:
        return 'Very high revenue country'
    else:
        return x
df['revenue_categories']=df.Revenue.apply(rev_categories)
```
