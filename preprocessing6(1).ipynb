```python
def clean_size(x):
    if type(x)=='float':
        return x
    elif x=='Varies with device':
        return x
    elif 'M' in x:
        return x
    elif 'k' in x:
        return str(round(float(x[:-1])/1024,2))+'M'
    else:
        return x
df['Size']=df['Size'].apply(clean_size)
```

```python
df['Installs'].unique()
```

```python
def clean_installs(x):

    if x in ['10+', '1+', '5+', '0+', '0', '50+', '100+', '500+']:
        return '< 1000'
    
    elif x in ['1,000+', '5,000+', '50,000+', '10,000+']:
        return '< 100000'
    
    elif x in ['100,000+', '500,000+']:
        return '< 1000000'
    
    else:
        return '>= 1000000'
df.Installs=df.Installs.apply(clean_installs)
df.Installs
```

## what is average numbers of reviews per content rating?

```python
# 🟢 تستخدم groupby إمتى؟

# لما تحب تجمّع البيانات حسب عمود categorical وتطبّق عليها إحصائية أو عملية حسابية على عمود numerical (أو حتى categorical).
reviews_per_content_=df.groupby('Content Rating')['Reviews'].mean().sort_values(ascending=True).reset_index()
reviews_per_content_
```

```python
px.bar(reviews_per_content_,x='Content Rating',y='Reviews',
      labels={'Reviews':'Avg_numbers_of_reveiws'},text_auto=True)
```

## what is average numbers of reviews per content rating per App Type?

```python
content_type_review=df.groupby(['Content Rating','Type'])['Reviews'].mean().reset_index()
content_type_review
```

```python
px.bar(content_type_review,x='Content Rating',y='Reviews',
      labels={'Reviews':'Avg_numbers_of_reveiws'},text_auto=True,color='Type',barmode='group')
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(drop='first',sparse_output=False)
ohe_train_arr=ohe.fit_transform(x_train[['Installs','Type','Content Rating']])
ohe_test_arr=ohe.transform(x_test[['Installs','Type','Content Rating']])
```

```python
ohe_train_df=pd.DataFrame(
    ohe_train_arr,columns=ohe.get_feature_names_out()
)
ohe_train_df
```

```python
ohe_test_df=pd.DataFrame(
    ohe_test_arr,columns=ohe.get_feature_names_out()
)
ohe_test_df
```
