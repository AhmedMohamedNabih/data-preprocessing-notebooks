```python
plt.plot(x,y,marker='^',c='g',linestyle='dotted',linewidth=2)
# plt.plot(x,y,z,y,marker='^',c='g',linestyle='dotted',linewidth=2)
# plt.plot(z,y,marker='o',c='r')
plt.xlabel('Feet',fontsize=15)
plt.ylabel('Price',fontsize=15)
plt.title('Feet VS Price')
plt.grid()
plt.show()
plt.plot(z,y,marker='o',c='r')
plt.xlabel('garageSize')
plt.ylabel('Price')
plt.title('GarageSize VS price')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,marker='^',c='g')
plt.xlabel('Feet',fontsize=15)
plt.ylabel('Price',fontsize=15)
plt.title('Feet VS Price')
plt.subplot(1,2,2)
plt.plot(z,y,marker='o',c='r')
plt.xlabel('garageSize',fontsize=13)
plt.ylabel('Price',fontsize=13)
plt.title('GarageSize VS price')
plt.tight_layout()
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,marker='^',c='g')
plt.xlabel('Feet',fontsize=15)
plt.ylabel('Price',fontsize=15)
plt.title('Feet VS Price')
plt.subplot(2,1,2)
plt.plot(z,y,marker='o',c='r')
plt.xlabel('garageSize',fontsize=13)
plt.ylabel('Price',fontsize=13)
plt.title('GarageSize VS price')
plt.tight_layout()
plt.show()
```

```python
x=np.array(['bmw','Mercedas','toyota','Porche','audi'])
y=np.array([10,5,20,5,7])
plt.bar(x,y,width=0.75)
plt.title('Car Sales')
plt.xlabel('Cars')
plt.ylabel('Sales')
plt.grid()
plt.show()
```

```python
# barh->horizantel
# barh -> مفيش فيها width -> لازمheight -> وهي هي نفس مهمه width
x=np.array(['bmw','Mercedas','toyota','Porche','audi'])
y=np.array([10,5,20,5,7])
plt.barh(x,y,height=0.75)
plt.title('Car Sales')
plt.xlabel('Cars')
plt.ylabel('Sales')
plt.grid()
plt.show()
```
