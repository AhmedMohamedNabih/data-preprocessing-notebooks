## Do developed and developing countries differ significantly in average life expectancy

```python
le_per_status=df.groupby('Status')['Life_expectancy'].mean().round(1).reset_index()
le_per_status
```

```python
px.bar(le_per_status,x='Status',y='Life_expectancy',labels={'Status':'country Type','Life_expectancy':'average Life_expectancy'})
```

## What is the average of Expenditure per status per year?
